```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creating a Standard plot with 2 different set's of data

danger_drop = [0,9,8,90,85,80,70,70,65,55,60,70,65,55]
railgun = [0, 75,70,60,65,60,45,55,50,40,40,35,35,30]
x_axis = np.arange(0,120,9)
y_axis = np.arange(0,100,9)
plt.plot(x_axis,railgun,marker='o',color='blue',label='RailGun')
plt.plot(x_axis,danger_drop,marker='x',color='orange',label='Danger Drop')
plt.ylabel('Speed (MPH)')
plt.xlabel('Time (Sec)')
plt.title('Different RollerCoasters and Their Speed')
plt.legend()
plt.grid()
plt.show()
```

```python
# Bar Chart

cities =['New Orleans','LA','New York','Houston','Las Vegas']
bars = [8.6,8.5,8.3,7.9,7.2]
x_axis = np.arange(len(bars))
plt.bar(x_axis,bars,tick_label=cities,color='red')
plt.xlabel('Cities')
plt.ylabel('Bars per 100,000 Households')
plt.ylim(0,10)
plt.title('Density of Bars in Cities')
plt.show()
```

```python
# Pie Chart

colors = ['yellow','blue','orange','red','purple','green','gray']
values = [17,23,45,67,34,98,78]
names = ['Peach','Pumpkin','Apple','Cherry','Pecan','Blueberry','Key Lime']
explode = (0,0,0,0,0,0,0.1)
plt.pie(values,explode=explode,labels=names,colors=colors,autopct='%1.1f%%',shadow=True,startangle=160)

plt.axis("equal")
```

```python
# Scatter Plots

sales = [200,225,350,325,400,425,430,400,450,500,550,600]
temp = [11,14,15,16,17,17.5,18,19,22,23,23.5,25]
plt.scatter(sales,temp,color='Orange',marker='v')
plt.xlabel('Number of Sales')
plt.ylabel('Temperature (C)')
plt.title('Number of Ice Cream Sales vs Temperature')
plt.grid()
```
