```python
data = pd.read_csv('movies.csv')
data.head()
```

```python
# Finding movies with greater than an IMDB rating of 7
print(data.loc[data['IMDB'] >= 7,['FILM','IMDB','IMDB_norm','IMDB_norm_round','IMDB_user_vote_count']].head())
```

```python
# Movies with Less than 20k votes and Higher than a rating of 7
print(data.loc[(data['IMDB_user_vote_count']<20000) & (data['IMDB']>7),['FILM','IMDB','IMDB_user_vote_count']].head())
```

```python
#Cleaning a Dataset
crime = pd.read_csv('crime.csv')
print(crime.count())
```

```python
crime = crime.dropna(how='any')
crime.count()
```

```python
# Condensing Offense Type to Common Categories
print(crime['Offense Type'].unique())
crime['Offense Type'] = crime['Offense Type'].replace({'Identity Theft':'Theft','Credit Card/ATM Fraud':'Fraud',
                                                       'Shoplifting':'Theft','Theft From Motor Vehicle':'Theft',
                                                       'Motor Vehicle Theft':'Theft','Burglary':'Theft',
                                                       'Theft of Motor Vehicle Parts or Accessories':'Theft',
                                                       'Theft From Building':'Theft','Robbery':'Theft',
                                                       'Theft From Coin-Operated Machine or Device':'Theft',
                                                       'Pocket-Picking':'Theft','Stolen Property Offenses':'Theft',
                                                       'Purse-Snatching':'Theft', 'Wire Fraud':'Fraud','Welfare Fraud':'Fraud',
                                                      'Drug/Narcotic Violations':'Violations','Weapons Law Violations':'Violations',
                                                      'Drug Equipment Violations':'Violations','Counterfeiting/Forgery':'Fraud',
                                                      'Prostitution':'Sex Crimes','Pornography/Obscene Material':'Sex Crimes',
                                                      'Assisting or Promoting Prostitution':'Sex Crimes','Commercial Sex Acts':'Sex Crimes',
                                                      'Impersonation':'Fraud'})
print('--------------------------------------------------------------------------------------------')
print(crime['Offense Type'].unique())
```

```python
# Creating a New Data Frame for a Specific Neighborhood
king_crime = crime.loc[crime['Neighborhood'] == 'King']
king_crime.head()
```
