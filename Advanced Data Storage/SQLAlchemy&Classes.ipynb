```python
from sqlalchemy import create_engine
import os

# Path to sqlite

file_path = os.path.abspath(os.getcwd())+"\Census_Data.db"
```

```python
engine = create_engine(f"sqlite:///{file_path}")
```

```python
# Query All Records in the the Database

data = engine.execute("SELECT * FROM Census_Data")

for record in data:
    print(record)
    break
```

```python
# Select City and States where the Population is Greater than 935,000 people

data = engine.execute("select city, state from Census_Data where Population > 935000")
for record in data:
    print(record)
```

```python
# Using Pandas and SQLAlchemy

import pandas as pd

conn = engine.connect()
data = pd.read_sql("select * from Census_Data",conn)
data.head()
```
