```python
from sqlalchemy import create_engine
from sqlalchemy import Table, Column, Integer, String
from flask import Flask
from flask_sqlalchemy import SQLAlchemy
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import func
```

```python
app = Flask(__name__)
app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///db/bigfoot.sqlite'
db = SQLAlchemy(app)

class bigfoot(db.Model):
    __tablename__ = 'bigfoot'
    
    id = db.Column(db.Integer, primary_key = True)
    number = db.Column(db.Integer)
    title = db.Column(db.String)
    classification = db.Column(db.String)
    timestamp = db.Column(db.String)
    latitude = db.Column(db.Float)
    longitude = db.Column(db.Float)
    
    def __repr__(self):
        return '<BigFoot %r>' % (self.name)
    
@app.before_first_request
def setup():
    db.drop_all()
    db.create_all()
```

```python
from datetime import datetime
import timestring
years = []

for val in bigfoot.query.all():
    year = timestring.Date(val.timestamp).year
    val.timestamp = year
    years.append(year)
  
```

```python
from sqlalchemy import func

results = db.session.query(bigfoot.id,func.count(bigfoot.timestamp)).group_by(bigfoot.timestamp).all()
results
```
